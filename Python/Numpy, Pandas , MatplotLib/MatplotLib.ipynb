```python
import matplotlib.pyplot as plt
```

# Line graph

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Y=X**2")
plt.savefig("Plot1.jpg")
```

# Bar graph

```python
plt.bar(x,y,color="black")
```

```python
plt.barh(x,y)
```

## Scatter plot

```python
plt.scatter(x,y,alpha=0.7,s=100, edgecolor="red")#alpha- thickness, s = size
```

## Histograms

```python
x=[2,2,3,4,5,5,6,7,3,5,6,7,4,3,5,6,7,3,2,5,6,7,8]
```

```python
plt.hist(x,alpha=0.4,color="red")
```

## PieChart

```python
plt.pie(x,labels=x)
plt.show()
```

## Box plot

```python
plt.boxplot(y)
plt.show()
```
